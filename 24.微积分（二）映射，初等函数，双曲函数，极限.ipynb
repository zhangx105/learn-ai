2.实数：自然数，整数，有理数，无理数。

实数包括有理数和无理数。

有理数包括所有整数和分数（q/p，其中p,q都是整数,p!=0）。

有理数的两个特点：有序性和稠密性；可以表示为有限小数，或者无限循环小数。

无理数：在数轴上出了有理点外，还有很多空隙，空隙处的点称为无理点；无理点是无限不循环小数。

3.函数。

偶函数和奇函数，偶函数是当定义域有x属于D(x),同时有-x属于D(x),这时有f(x)=f(-x), 
如y = x的2次方:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math

x = range(-10, 11)
y = np.square(x)
plt.plot(x, y)
plt.show()
```

以上是偶函数，图形关于纵坐标对称。

奇函数的例子，y=sin(x)：

```python
x = np.arange(-10, 11, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

sin函数是奇函数，关于原点对称。

单调增加函数，sin函数在 -pi/2 - pi/2 上是单调增加函数：

```python
x = np.arange(-math.pi/2, math.pi/2, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

单调减少函数，cos函数在 0 - pi 上是单调减少函数：

```python
x = np.arange(0, math.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.show()
```

幂函数y=x**n，当n>0时，在（0, + ∞）上单调增加；当n<0时，在（0, + ∞）上单调减少，
如下图所示：

```python
x = np.arange(0, 10, 0.1)
y = x**3
plt.plot(x, y)
plt.show()
```

反函数：函数的自变量和因变量交换后得到的新函数，叫反函数。函数和它的反函数关于直线
y=x对称。如y=x^2 和y=x^1/2互为反函数:

```python
x = np.arange(0, 10, 0.1)
y = x**2
plt.plot(x, y)

x = np.arange(0, 80, 1)
y = x**0.5
plt.plot(x, y)

plt.plot([0, 80], [0, 80])
plt.axis([0, 100, 0, 100])

plt.show()
```

复合函数是将一个函数代入另一个函数而得到的新函数。

3.映射。

设X, Y 是两个非空集合, 如果按照某种确定的
法则f , 使得对于每个x ∈ X, 都有唯一的一个y ∈ Y与其对应, 则
称这个对应关系f 为定义于X 取值于Y 的一个映射(mapping) , 并
且记之为
f : X → Y.

包括单射，满射，双射，逆映射等。

单射：由x 1 ≠ x 2 可以推出f ( x 1 ) ≠ f ( x 2 )

函数和反函数是逆映射的例子。

4.初等函数和非初等函数。

初等函数包括常数函数，幂函数，指数函数，对数函数，三角函数，反三角函数等。

常数函数，y=C，C是常数，图形上是一条直线，如y=30：

```python
x = [1,2,3,4,5]
y = [30,30,30,30,30]
plt.plot(x, y)

plt.show()
```

幂函数，y=x^n，n为任意常数，该曲线通过平面上的点(1 ，1)，如y=x^2：

```python
x = np.arange(0,10, 0.1)
y = x**2
plt.plot(x, y)

plt.show()
```

指数函数，y=a^x，a为任意正数，如y=3^x:

```python
x = np.arange(0,10, 0.1)
y = 3**x
plt.plot(x, y)

plt.show()
```

对数函数，y = loga x，当a > 1 时, 对数函数
的定义域是( 0, + ∞) , 在整个定义域内, 对数函数是单调增加的；如y=log10 x,

```python
x = np.arange(1,10, 0.1)
y = np.log10(x)
plt.plot(x, y)

plt.show()
```

三角函数，包括sinx,cosx,tanx,cotx,secx,cscx等，如下所示：

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title("y=sin(x)")

plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.title("y=cos(x)")

plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.tan(x)
plt.plot(x, y)
plt.title("y=tan(x)")

plt.show()
```
